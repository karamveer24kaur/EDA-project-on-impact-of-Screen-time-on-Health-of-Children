```python
# Splitting Health Impacts Column into 4 different columns 
# headings = names of potential health risks
# values = yes/no values for each row

# str.split function = to split the string into a list of impacts
# .apply function = to create a new column for each impact 

# to change the dtype of health_impacts column from float to str, as split method works on str and not on float
df['Health_Impacts'] = df['Health_Impacts'].astype(str)

df['Poor Sleep'] = df['Health_Impacts'].apply(lambda x: 'Poor Sleep' in x.split(', '))
df['Anxiety'] = df['Health_Impacts'].apply(lambda x: 'Anxiety' in x.split(', '))
df['Obesity Risk'] = df['Health_Impacts'].apply(lambda x: 'Obesity Risk' in x.split(', '))
df['Eye Strain'] = df['Health_Impacts'].apply(lambda x: 'Eye Strain' in x.split(', '))
```

# Bivariate Analysis (Relationships Between Two Variables)
# Objective: Explore relationships and correlations between pairs of variables.

```python
# 1.) Screen Time vs. Age 

plt.figure(figsize=(12, 7))
sns.boxplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, palette='coolwarm')

plt.title('Average Daily Screen Time by Age')
plt.xlabel('Age (Years)')
plt.ylabel('Average Daily Screen Time (hours)')

plt.show()
```

# Output Interpretation:

- Screen Time vs. Age: The box plots shows a trend where Avg_Daily_Screen_Time_hr tends to increase with age, especially in the teenage years (e.g., 13-18). 

- Younger children (8-10) have lower median screen times. This is a common pattern as older children gain more autonomy and access to devices.

```python
# 2.) Screen Time vs. Gender

plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df, palette='pastel')

plt.title('Average Daily Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Daily Screen Time (hours)')

plt.show()
```

# Output Interpretation:

- Screen Time vs. Gender: The Avg_Daily_Screen_Time_hr is almost similar between genders

```python
# 3.) Screen Time vs. Primary Device

plt.figure(figsize=(12, 7))
sns.boxplot(x='Primary_Device', y='Avg_Daily_Screen_Time_hr', data=df, palette='viridis',
            order=df.groupby('Primary_Device')['Avg_Daily_Screen_Time_hr'].median().sort_values(ascending=False).index)

plt.title('Average Daily Screen Time by Primary Device')
plt.xlabel('Primary Device')
plt.ylabel('Average Daily Screen Time (hours)')

plt.show()
```

# Output Interpretation:

- Screen Time vs. Primary Device: 'Smartphone' has highest average screen time, as it is portable and used for various activities. 

```python
# 4.) Exceeded Limit vs. Age/Gender/Device 

plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='Exceeded_Recommended_Limit', data=df, palette='Blues')

plt.title('Proportion of Children Exceeding Limit by Age')
plt.xlabel('Age (Years)')
plt.ylabel('Proportion Exceeding Limit (True=1, False=0)')

plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Exceeded_Recommended_Limit', data=df, palette='Greens')

plt.title('Proportion of Children Exceeding Limit by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion Exceeding Limit (True=1, False=0)')

plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Primary_Device', y='Exceeded_Recommended_Limit', data=df, palette='Oranges',
            order=df.groupby('Primary_Device')['Exceeded_Recommended_Limit'].mean().sort_values(ascending=False).index)

plt.title('Proportion of Children Exceeding Limit by Primary Device')
plt.xlabel('Primary Device')
plt.ylabel('Proportion Exceeding Limit (True=1, False=0)')

plt.show()
```

# Output Interpretation:

- Exceeded Limit vs. Age/Gender/Device:

- By Age: The proportion exceeding the limit increases with age, mirroring the screen time trend. 

- Older children are more likely to exceed the limit.

- By Gender: Similar to average screen time, the proportion exceeding the limit does not show a drastic difference between genders.

- By Primary Device: Children whose primary device is a 'Smartphone', have a higher proportion exceeding the recommended limit.

```python
# 5.) Health Impacts vs. Screen Time (Average Screen Time for each impact)

plt.figure(figsize=(14, 7))
sns.barplot(x=['Poor Sleep', 'Anxiety', 'Obesity Risk', 'Eye Strain'], 
            y=[df['Poor Sleep'].mean(), df['Anxiety'].mean(), df['Obesity Risk'].mean(), df['Eye Strain'].mean()])

plt.title('Average Daily Screen Time by Reported Health Impact')
plt.xlabel('Health Impact')
plt.ylabel('Average Daily Screen Time (hours)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() #  to automatically adjust the spacing between plot elements) to prevent overlapping labels, titles, and ticks

plt.show()
```

# Output Interpretation:

- Health Impacts vs. Screen Time: Impacts like 'Poor Sleep', 'Eye Strain' are associated with significantly higher average daily screen times compared to other impacts.

- This strongly suggests a correlation between prolonged screen exposure and negative health outcomes.

# Multivariate Analysis (Relationships Among Multiple Variables)
# Objective: Discover more complex patterns and interactions.

```python
# 1.) Correlation Matrix for Numerical Variables 

numerical_cols = ['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')

plt.show()
```

# Output Interpretation:

- Age and Avg_Daily_Screen_Time_hr show a positive correlation, indicating that as age increases, screen time tends to increase.

- Educational_to_Recreational_Ratio hasa weak or no significant correlation with Age or Avg_Daily_Screen_Time_hr. 

- This suggests that simply spending more time on screens or being older doesn't necessarily dictate the quality of that screen time.

```python
# 2.) Educational Ratio vs. Screen Time, colored by Exceeded Limit

plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='Avg_Daily_Screen_Time_hr',
    y='Educational_to_Recreational_Ratio',
    hue='Exceeded_Recommended_Limit',
    style='Urban_or_Rural', # Add another dimension
    palette='coolwarm',
    alpha=0.7
)

plt.title('Educational Ratio vs. Screen Time, by Exceeded Limit and Location')
plt.xlabel('Average Daily Screen Time (hours)')
plt.ylabel('Educational to Recreational Ratio')
plt.legend(title='Exceeded Limit / Location')

plt.show()
```

# Output Interpretation: 

- This scatter plot helps visualize if there's a "sweet spot" for the educational ratio.

- Or if exceeding the limit is purely about duration regardless of content.

- It is seen that points representing Exceeded_Recommended_Limit = True are generally clustered towards higher Avg_Daily_Screen_Time_hr values regardless of their Educational_to_Recreational_Ratio. 
    
- This would suggest that the duration of screen time is a primary driver for exceeding limits, rather than just the type of content.
