```python
# For Health_Impacts, it's multi-label, so we'll handle it separately later.
# For now, just show its top few values
print(df['Health_Impacts'].value_counts().head(10))
```

```python
# Splitting Health Impacts Column into 4 different columns 
# headings = names of potential health risks
# values = yes/no values for each row

# str.split function = to split the string into a list of impacts
# .apply function = to create a new column for each impact 

# to change the dtype of health_impacts column from float to str, as split method works on str and not on float
df['Health_Impacts'] = df['Health_Impacts'].astype(str)

df['Poor Sleep'] = df['Health_Impacts'].apply(lambda x: 'Poor Sleep' in x.split(', '))
df['Anxiety'] = df['Health_Impacts'].apply(lambda x: 'Anxiety' in x.split(', '))
df['Obesity Risk'] = df['Health_Impacts'].apply(lambda x: 'Obesity Risk' in x.split(', '))
df['Eye Strain'] = df['Health_Impacts'].apply(lambda x: 'Eye Strain' in x.split(', '))
```

```python
# Summary of Initial Data Exploration:

# The dataset contains 9712 entries and 8 columns, with no missing values.
    
# The data types are int64 for Age, float64 for Avg_Daily_Screen_Time_hr and Educational_to_Recreational_Ratio, object for Gender, 
# Primary_Device, Health_Impacts, and Urban_or_Rural.
    
# The mean age is 12.98 years, and the mean average daily screen time is 4.35 hours.
    
# The standard deviation for age is 3.16 years, and for average daily screen time is 1.72 hours.
    
# The minimum age is 8 years, and the maximum age is 18 years.
    
# The minimum average daily screen time is 0 hours, and the maximum average daily screen time is 13.89 hours.
    
# The 25th percentile age is 10 years, and the 50th percentile average daily screen time is 4.44 hours.
    
# The 75th percentile age is 16 years, and the 75th percentile average daily screen time is 5.38 hours.
    
# The maximum age is 18 years, and the maximum average daily screen time is 13.89 hours.
    
# The dataset is predominantly male (4942) and female (4770), with a slight bias towards males.
    
# The primary devices used are smartphones (4568), TVs (2487), laptops (1433), and tablets (1224).
    
# The majority of children (8301) exceeded the recommended screen time limit.
    
# The urban_or_rural distribution is 6851 urban and 2861 rural.
    
# Key Findings and Recommendations:

# High Screen Time Prevalence: A significant majority of children (8301) exceeded the recommended screen time limit, indicating a widespread issue.
    
# Age-Related Increase: Average daily screen time and the likelihood of exceeding limits tend to increase with age, peaking in the late teenage years.
    
# Device Dependency: Smartphones and laptops are the primary drivers of high screen time and exceeding limits, likely due to their versatility 
# and constant accessibility.
    
# Direct Link to Health Impacts: There's a strong correlation between exceeding recommended screen time limits and the occurrence of negative 
# health impacts. 
    
# 'Poor Sleep', 'Eye Strain', and 'Anxiety' are consistently the most reported issues.
    
# Educational Content Not a Panacea: While educational content is present, a higher educational-to-recreational ratio doesn't necessarily 
# prevent children from exceeding overall screen time limits or experiencing negative health impacts, 
# suggesting that total duration is a critical factor.
    
# Urban vs. Rural Differences: Minor differences might exist, but the core trends of screen time and health impacts are likely consistent 
# across urban and rural settings.
```

```python
# 1.) Age Distribution 

plt.figure(figsize=(10, 6)) # make a figure first, specify its dimensions
sns.histplot(df['Age'], bins=11, kde=True, color='skyblue') # make a histplot of age column

# add title, labels, etc.
plt.title('Distribution of Age') 
plt.xlabel('Age (Years)')
plt.ylabel('Number of Children')
plt.xticks(range(8, 19)) # to set the x-axis tick marks on a plot

plt.show()
```

```python
# Output Interpretation

# Age Distribution: The age distribution appears relatively uniform across the 8-18 age range, with perhaps slight variations depending on 
# the random data generation. This suggests a good spread of age groups in the dataset.
```

```python
# 2.) Gender Distribution 

plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df, palette='viridis') # used when there are discrete categories

plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Children')

plt.show()
```

```python
# Gender Distribution: As noted in the initial inspection, gender is almost perfectly balanced, indicating no significant bias in representation.
```

```python
# Average Daily Screen Time:
# The histogram shows a wide distribution of screen times, with a peak around the mean (4.5 hours).
# The box plot confirms the spread and might show some outliers at the higher end, indicating some children have exceptionally long screen times.
```

```python
# 1.) Screen Time vs. Age 

plt.figure(figsize=(12, 7))
sns.boxplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, palette='coolwarm')

plt.title('Average Daily Screen Time by Age')
plt.xlabel('Age (Years)')
plt.ylabel('Average Daily Screen Time (hours)')

plt.show()
```

```python
#  Output Interpretation:

# Screen Time vs. Age:
# The box plots shows a trend where Avg_Daily_Screen_Time_hr tends to increase with age, especially in the teenage years (e.g., 13-18). 
# Younger children (8-10) have lower median screen times. This is a common pattern as older children gain more autonomy and access to devices.
```

```python
# 2.) Screen Time vs. Gender

plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df, palette='pastel')

plt.title('Average Daily Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Daily Screen Time (hours)')

plt.show()
```

```python
# Screen Time vs. Gender:
# The Avg_Daily_Screen_Time_hr is almost similar between genders
```

```python
# 3.) Screen Time vs. Primary Device

plt.figure(figsize=(12, 7))
sns.boxplot(x='Primary_Device', y='Avg_Daily_Screen_Time_hr', data=df, palette='viridis',
            order=df.groupby('Primary_Device')['Avg_Daily_Screen_Time_hr'].median().sort_values(ascending=False).index)

plt.title('Average Daily Screen Time by Primary Device')
plt.xlabel('Primary Device')
plt.ylabel('Average Daily Screen Time (hours)')

plt.show()
```

```python
# Screen Time vs. Primary Device:
# 'Smartphone' has highest average screen time, as it is portable and used for various activities. 
```

```python
# 4.) Exceeded Limit vs. Age/Gender/Device 

plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='Exceeded_Recommended_Limit', data=df, palette='Blues')

plt.title('Proportion of Children Exceeding Limit by Age')
plt.xlabel('Age (Years)')
plt.ylabel('Proportion Exceeding Limit (True=1, False=0)')

plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Exceeded_Recommended_Limit', data=df, palette='Greens')

plt.title('Proportion of Children Exceeding Limit by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion Exceeding Limit (True=1, False=0)')

plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Primary_Device', y='Exceeded_Recommended_Limit', data=df, palette='Oranges',
            order=df.groupby('Primary_Device')['Exceeded_Recommended_Limit'].mean().sort_values(ascending=False).index)

plt.title('Proportion of Children Exceeding Limit by Primary Device')
plt.xlabel('Primary Device')
plt.ylabel('Proportion Exceeding Limit (True=1, False=0)')

plt.show()
```

```python
# Exceeded Limit vs. Age/Gender/Device:

# By Age: The proportion exceeding the limit increases with age, mirroring the screen time trend. 
# Older children are more likely to exceed the limit.
# By Gender: Similar to average screen time, the proportion exceeding the limit does not show a drastic difference between genders.
# By Primary Device: Children whose primary device is a 'Smartphone', have a higher proportion exceeding the recommended limit.
```

```python
# 5.) Health Impacts vs. Screen Time (Average Screen Time for each impact)

plt.figure(figsize=(14, 7))
sns.barplot(x=['Poor Sleep', 'Anxiety', 'Obesity Risk', 'Eye Strain'], 
            y=[df['Poor Sleep'].mean(), df['Anxiety'].mean(), df['Obesity Risk'].mean(), df['Eye Strain'].mean()])

plt.title('Average Daily Screen Time by Reported Health Impact')
plt.xlabel('Health Impact')
plt.ylabel('Average Daily Screen Time (hours)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() #  to automatically adjust the spacing between plot elements) to prevent overlapping labels, titles, and ticks

plt.show()
```

```python
# Health Impacts vs. Screen Time:

# Impacts like 'Poor Sleep', 'Eye Strain' are associated with significantly higher average daily screen times compared to other impacts. 
# This strongly suggests a correlation between prolonged screen exposure and negative health outcomes.
```

```python
# 6.) Educational Ratio vs. Screen Time 

plt.figure(figsize=(10, 7))
sns.scatterplot(x='Avg_Daily_Screen_Time_hr', y='Educational_to_Recreational_Ratio', data=df, alpha=0.6, color='purple')

plt.title('Educational to Recreational Ratio vs. Average Daily Screen Time')
plt.xlabel('Average Daily Screen Time (hours)')
plt.ylabel('Educational to Recreational Ratio')

plt.show()
```

```python
# Educational Ratio vs. Screen Time:

# The scatter plot shows no clear linear correlation. 
# high screen time doesn't necessarily mean low educational ratio, or vice-versa. 
# Some children have high screen time with a good balance of educational content.
# while others might have low screen time but still mostly recreational.
```

```python
# 7.) Urban/Rural vs. Screen Time 

plt.figure(figsize=(8, 6))
sns.boxplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df, palette='tab10')

plt.title('Average Daily Screen Time by Urban or Rural Location')
plt.xlabel('Location Type')
plt.ylabel('Average Daily Screen Time (hours)')

plt.show()
```

```python
# Urban/Rural vs. Screen Time

# There is a subtle differences. 
# For instance, urban children have slightly higher screen times due to greater access to technology.
```

```python
# Multivariate Analysis (Relationships Among Multiple Variables)
# Objective: Discover more complex patterns and interactions.
```

```python
# 1.) Correlation Matrix for Numerical Variables 

numerical_cols = ['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')

plt.show()
```

```python
# Output analysis

# Age and Avg_Daily_Screen_Time_hr show a positive correlation, indicating that as age increases, screen time tends to increase.
# Educational_to_Recreational_Ratio hasa weak or no significant correlation with Age or Avg_Daily_Screen_Time_hr. 
# This suggests that simply spending more time on screens or being older doesn't necessarily dictate the quality of that screen time.
```

```python
# 2.) Educational Ratio vs. Screen Time, colored by Exceeded Limit

plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='Avg_Daily_Screen_Time_hr',
    y='Educational_to_Recreational_Ratio',
    hue='Exceeded_Recommended_Limit',
    style='Urban_or_Rural', # Add another dimension
    palette='coolwarm',
    alpha=0.7
)

plt.title('Educational Ratio vs. Screen Time, by Exceeded Limit and Location')
plt.xlabel('Average Daily Screen Time (hours)')
plt.ylabel('Educational to Recreational Ratio')
plt.legend(title='Exceeded Limit / Location')

plt.show()
```

```python
# output analysis 

# This scatter plot helps visualize if there's a "sweet spot" for the educational ratio.
# Or if exceeding the limit is purely about duration regardless of content.
# it is seen that points representing Exceeded_Recommended_Limit = True are generally clustered towards higher Avg_Daily_Screen_Time_hr values
# regardless of their Educational_to_Recreational_Ratio. 
    
# This would suggest that the duration of screen time is a primary driver for exceeding limits, rather than just the type of content.
```
